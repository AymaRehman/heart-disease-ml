```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
sns.set_theme(style="whitegrid", palette="Set2", font_scale=1.1)
CLASS_COLORS = {0: "#2a9d8f", 1: "#c1121f"}
CLASS_LABELS = {0: "No Disease (0)", 1: "Disease Present (1)"}

print("Libraries loaded successfully.")
```

### REQUIREMENT:
"a snippet of the data file structure showing all columns of the data file
and their values for at least some data objects"

```python
df = pd.read_csv("../data/heart_cleaned.csv")

print(f"Rows    : {len(df)}")
print(f"Columns : {len(df.columns)}")
print(f"Features: {list(df.drop(columns='target').columns)}")
print("Target  : 'target' (0 = No Disease, 1 = Disease Present)")
print()
print("First 5 rows:")
display(df.head())
```

### REQUIREMENT:
"Are the classes in the dataset balanced, or does one class (or several
classes) prevail? It is determined by judging how many data objects belong
to each class."

```python
class_counts = df["target"].value_counts().sort_index()
class_pct    = df["target"].value_counts(normalize=True).sort_index() * 100

balance_df = pd.DataFrame({
    "Class Label": ["0 — No Disease", "1 — Disease Present"],
    "Count": class_counts.values,
    "Percentage (%)": class_pct.round(1).values
})

print("Class Distribution:")
display(balance_df.set_index("Class Label"))

# Visual bar chart of class balance
fig, ax = plt.subplots(figsize=(5, 4))
bars = ax.bar(
    ["No Disease (0)", "Disease Present (1)"],
    class_counts.values,
    color=[CLASS_COLORS[0], CLASS_COLORS[1]],
    edgecolor="white", linewidth=1.5
)
for bar, pct in zip(bars, class_pct.values):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 2,
        f"{pct:.1f}%",
        ha="center", va="bottom", fontweight="bold"
    )
ax.set_title("Figure 1. Class Distribution", fontweight="bold")
ax.set_ylabel("Number of Data Objects")
ax.set_ylim(0, max(class_counts.values) * 1.15)
plt.tight_layout()
plt.show()
```

### REQUIREMENT:
"create at least two 2- or 3-dimensional scatter plots (widget Scatter Plot)
illustrating class separability based on different features (attributes);
students should avoid using the data object ID or class label as a variable
in the scatter plots"

```python
# Scatter Plot 1: Age (age) vs Maximum Heart Rate (thalach) 

fig, ax = plt.subplots(figsize=(8, 5))

for label, color in CLASS_COLORS.items():
    subset = df[df["target"] == label]
    ax.scatter(
        subset["age"], subset["thalach"],
        c=color, label=CLASS_LABELS[label],
        alpha=0.65, edgecolors="white", linewidth=0.4, s=60
    )

ax.set_xlabel("Age (years)", fontweight="bold")
ax.set_ylabel("Maximum Heart Rate Achieved - thalach (bpm)", fontweight="bold")
ax.set_title("Figure 2. Scatter Plot 1: Age vs Maximum Heart Rate by Class", fontweight="bold")
ax.legend(title="Target Class", framealpha=0.9)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot 2: ST Depression (oldpeak) vs Cholesterol (chol) 
fig, ax = plt.subplots(figsize=(8, 5))

for label, color in CLASS_COLORS.items():
    subset = df[df["target"] == label]
    ax.scatter(
        subset["oldpeak"], subset["chol"],
        c=color, label=CLASS_LABELS[label],
        alpha=0.65, edgecolors="white", linewidth=0.4, s=60
    )

ax.set_xlabel("ST Depression Induced by Exercise - oldpeak", fontweight="bold")
ax.set_ylabel("Serum Cholesterol - chol (mg/dl)", fontweight="bold")
ax.set_title("Figure 3. Scatter Plot 2: ST Depression vs Cholesterol by Class", fontweight="bold")
ax.legend(title="Target Class", framealpha=0.9)
plt.tight_layout()
plt.show()
```

### REQUIREMENT:
"create at least 2 histograms (widget Bar Plot) showing the separation of
classes based on features (attributes) of interest"


```python
# Histogram 1: Maximum Heart Rate (thalach) by Class 
fig, ax = plt.subplots(figsize=(8, 5))

for label, color in CLASS_COLORS.items():
    subset = df[df["target"] == label]
    ax.hist(
        subset["thalach"], bins=18, alpha=0.65,
        color=color, label=CLASS_LABELS[label],
        edgecolor="white", linewidth=0.6
    )

for label, color in CLASS_COLORS.items():
    mean_val = df[df["target"] == label]["thalach"].mean()
    ax.axvline(mean_val, color=color, linestyle="--", linewidth=2,
               label=f"Mean (class {label}): {mean_val:.1f}")

ax.set_xlabel("Maximum Heart Rate Achieved - thalach (bpm)", fontweight="bold")
ax.set_ylabel("Frequency", fontweight="bold")
ax.set_title("Figure 4. Histogram 1: Maximum Heart Rate by Class", fontweight="bold")
ax.legend(framealpha=0.9)
plt.tight_layout()
plt.show()
```

```python
# Histogram 2: ST Depression (oldpeak) by Class 
fig, ax = plt.subplots(figsize=(8, 5))

for label, color in CLASS_COLORS.items():
    subset = df[df["target"] == label]
    ax.hist(
        subset["oldpeak"], bins=18, alpha=0.65,
        color=color, label=CLASS_LABELS[label],
        edgecolor="white", linewidth=0.6
    )

for label, color in CLASS_COLORS.items():
    mean_val = df[df["target"] == label]["oldpeak"].mean()
    ax.axvline(mean_val, color=color, linestyle="--", linewidth=2,
               label=f"Mean (class {label}): {mean_val:.2f}")

ax.set_xlabel("ST Depression Induced by Exercise - oldpeak", fontweight="bold")
ax.set_ylabel("Frequency", fontweight="bold")
ax.set_title("Figure 5. Histogram 2: ST Depression (oldpeak) by Class", fontweight="bold")
ax.legend(framealpha=0.9)
plt.tight_layout()
plt.show()
```

### REQUIREMENT:
"show the distribution of the 2 features (attributes) of interest
(widgets Distributions, Box Plot, Violin Plot)"


```python
# Box Plot: Maximum Heart Rate (thalach) vs Class
fig, ax = plt.subplots(figsize=(6, 5))

groups = [df[df["target"] == label]["thalach"].values for label in [0, 1]]
bp = ax.boxplot(
    groups,
    tick_labels=["No Disease (0)", "Disease Present (1)"],
    patch_artist=True,
    medianprops=dict(color="black", linewidth=2),
    flierprops=dict(marker="o", markerfacecolor="grey", markersize=4, alpha=0.5)
)

for patch, color in zip(bp["boxes"], [CLASS_COLORS[0], CLASS_COLORS[1]]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_title("Figure 6. Box Plot: Maximum Heart Rate (thalach) by Class", fontweight="bold")
ax.set_ylabel("Maximum Heart Rate Achieved (bpm)", fontweight="bold")
ax.set_xlabel("Target Class", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# Violin Plot: ST Depression (oldpeak) vs Class 
fig, ax = plt.subplots(figsize=(6, 5))

sns.violinplot(
    data=df,
    x="target", 
    y="oldpeak", 
    hue="target",          
    palette=CLASS_COLORS, 
    inner="box", 
    legend=False,          
    ax=ax
)

ax.set_xticks([0, 1])
ax.set_xticklabels(["No Disease (0)", "Disease Present (1)"])
ax.set_title("Figure 7. Violin Plot: ST Depression (oldpeak) by Class", fontweight="bold")
ax.set_ylabel("ST Depression Induced by Exercise - oldpeak", fontweight="bold")
ax.set_xlabel("Target Class", fontweight="bold")
plt.tight_layout()
plt.show()
```

### REQUIREMENT:
"calculate statistical indicators (at least the central tendency and the
dispersion of the feature values) (widget Column Statistics)"

### REQUIREMENT:
"conclusions from the analysis of statistical indicators (at least the
central tendency and the dispersion of the feature values)"

```python
features = [c for c in df.columns if c != "target"]
feature_df = df[features]

stats = pd.DataFrame(index=features)

# Central tendency
stats["Mean"]     = feature_df.mean().round(3)
stats["Median"]   = feature_df.median().round(3)

# Dispersion
stats["Std Dev"]  = feature_df.std().round(3)
stats["Variance"] = feature_df.var().round(3)
stats["Min"]      = feature_df.min().round(3)
stats["Max"]      = feature_df.max().round(3)
stats["Q1 (25%)"] = feature_df.quantile(0.25).round(3)
stats["Q3 (75%)"] = feature_df.quantile(0.75).round(3)
stats["IQR"]      = (feature_df.quantile(0.75) - feature_df.quantile(0.25)).round(3)

# Shape
stats["Skewness"] = feature_df.skew().round(3)
stats["Kurtosis"] = feature_df.kurt().round(3)

print("Table 1. Statistical Summary of All Features (cleaned dataset)")
display(
    stats.style
    .set_caption("Statistical Summary - Heart Disease Dataset (cleaned)")
    .format(precision=3)
    .set_table_styles([{
        "selector": "th",
        "props": [("background-color", "#4C9BE8"), ("color", "white"), ("font-weight", "bold")]
    }])
)

# Per-class statistics for the two most discriminative continuous features
print("Table 2. Per class Descriptive Statistics for thalach and oldpeak")
print()
for feat in ["thalach", "oldpeak"]:
    print(f"Feature: {feat}")
    per_class = df.groupby("target")[feat].agg(
        Mean="mean", Median="median",
        Std="std", Variance="var",
        Min="min", Max="max"
    ).round(3)
    per_class.index = [CLASS_LABELS[i] for i in per_class.index]
    display(per_class)
    print()
```
